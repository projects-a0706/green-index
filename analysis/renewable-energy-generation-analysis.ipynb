```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Added in Excel column 'population' from CO2 dataset and values 'per capita' - (Wh) per person, 1 TWh = 10**9

df = pd.read_csv('../data_final/renewable-energy-generation-population.csv', sep=';')

# Rename columns
filtered_df = df.rename(columns={'Entity': 'Country',
                                'year': 'Year',
                                'Other renewables (including geothermal and biomass) electricity generation - TWh':'Other renewables (TWh)',
                                'Solar generation - TWh': 'Solar generation (per capita - Wh)',
                                'Wind generation - TWh': 'Wind generation (TWh)',
                                'Hydro generation - TWh': 'Hydro generation (TWh)',
                                })

# Sort
filtered_df = filtered_df.sort_values(by='Country')
filtered_df.head(20)
```

```python
columns_to_convert = [
    'Other renewables (TWh)',
    'Solar generation (per capita - Wh)',
    'Wind generation (TWh)',
    'Hydro generation (TWh)',
    'Other renewable (per capita - Wh)',
    'Solar generation (per capita - Wh)',
    'Wind generation (per capita - Wh)',
    'Hydro generation (per capita - Wh)',
    'Population'
]

# Replace , mit .
for col in columns_to_convert:
    filtered_df[col] = (
        filtered_df[col]
        .astype(str)
        .str.replace(',', '.')  # remove commas
        .str.strip()            # remove whitespace
    )

print(f'Columns type: {filtered_df['Other renewable (per capita - Wh)'].dtype}')

# Convert to numeric
filtered_df[columns_to_convert] = filtered_df[columns_to_convert].apply(pd.to_numeric, errors='coerce')

filtered_df.head(20)
```

```python
print(f'Data shape: {filtered_df.shape[0]} rows, {filtered_df.shape[1]} columns')
```

```python
# Get the distinct countries
distinct_countries_count = filtered_df['Country'].nunique()
distinct_countries = filtered_df['Country'].unique()

# Display the result
print(f'There are {distinct_countries_count} countries in the dataset')
print(f'Countries: {distinct_countries}')

```

```python
# Rename countries to match the names from other datasets
country_mapping = {
    'Czechia': 'Czech Republic',
    'Russia': 'Russian Federation'
}

filtered_df['Country'] = filtered_df['Country'].replace(country_mapping)
```

```python
# Get the distinct countries
distinct_countries = filtered_df['Country'].unique()

# Display the result
print(f'Countries: {distinct_countries}')
```

```python
# Renewable energy generation - higher % of -> better sustainability 
# Hydro generation - TWh
avg_renewable_hydro = filtered_df.groupby('Country')['Hydro generation (per capita - Wh)'].mean().reset_index()

max_value = avg_renewable_hydro['Hydro generation (per capita - Wh)'].max()
min_value = avg_renewable_hydro['Hydro generation (per capita - Wh)'].min()

avg_renewable_hydro['renewable_hydro_generation_subindex'] = (avg_renewable_hydro['Hydro generation (per capita - Wh)'] - min_value)/(max_value - min_value)

avg_renewable_hydro.sort_values(by='renewable_hydro_generation_subindex', ascending=False)
```

```python
avg_renewable_hydro.to_csv('../data_subindex/subindex_renewable_hydro_energy_generation_2019-2023.csv', index=False)
```

```python
avg_renewable_hydro = avg_renewable_hydro.sort_values(by='renewable_hydro_generation_subindex')

plt.figure(figsize=(10,15))
plt.barh(avg_renewable_hydro['Country'],avg_renewable_hydro['renewable_hydro_generation_subindex'], color='green')
plt.xlabel('0 = worst, 1 = best')
plt.title('Renewable hydro energy generation subindex')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
